from bisect import bisect_right

class Interpolate:
    def __init__(self, x_list, y_list):
        if any(y - x <= 0 for x, y in zip(x_list, x_list[1:])):
            raise ValueError("x_list must be in strictly ascending order!")
        self.x_list = x_list
        self.y_list = y_list
        intervals = zip(x_list, x_list[1:], y_list, y_list[1:])
        self.slopes = [(y2 - y1) / (x2 - x1) for x1, x2, y1, y2 in intervals]

    def __call__(self, x):
        if not (self.x_list[0] <= x <= self.x_list[-1]):
            raise ValueError("x out of bounds!")
        if x == self.x_list[-1]:
            return self.y_list[-1]
        i = bisect_right(self.x_list, x) - 1
        return self.y_list[i] + self.slopes[i] * (x - self.x_list[i])

```python
for data_file in file:
    MOKE_data = pd.read_csv('MOKE_FeCoGd20221025C_Ta1_Fe100Co15Gd50_1m15s_Run1_noPI2022_10_25_time_12_26_59.csv')
    # finding the max field value (MOKE data would be min field since start with neg )
    xmax=(MOKE_data['Voltage (V)'].max()) 
    #print (xmax)

    xlast=(MOKE_data['Voltage (V)'].iloc[-1])# this is the last field point in +field 
    #print  (xlast)

    xtol=np.arange(xmax*0.95,xmax*1.05,dtype="int")
    #print  (xtol)
    file_title = data_file.replace(".DAT_VolMag_data.csv", "")

    #A=PPMS_data['Moment (emu/ccm)'].nsmallest(2)
    
    x = MOKE_data["Voltage (V)"]
    y = MOKE_data['MOKE Signal']

    #Plot raw data
    #file_title = data_file.replace(".DAT","")

    fig1, ax = plt.subplots(2,2, figsize = (16,10))

    fig1.suptitle(file_title, y = 1.02)
    ax[0,0].set_title("Raw Data")
    ax[0,0].plot(x, y)
    ax[0,0].grid()
    ax[0,0].set_xlabel("Voltage (V)")
    ax[0,0].set_ylabel("MOKE Signal")


    """To remove virgin curve we first need to find the last index point. Since data goes from 0 oe to max, from max 
    to min, min back to max we will sort field from highest to lowest and keep only the first 3 since those should
    corresponsd to max field. Of those 3 the one with the lowest index will be the end of the virgin curve."""

    PPMS_data_red = MOKE_data

    PPMS_data_red = PPMS_data_red.reset_index(drop = True)#Reset indexing so it starts at zero again.
    #We want to fit the positive to negative field region so this corresponds to the first half of the full loop.
    half_loop = int(len(PPMS_data_red)/2) 

    x_axis = PPMS_data_red["Voltage (V)"]
    x_labs = "Voltage (V)"

   
    """ To background correct we need to fit a line to one of the end points and subtract from the PPMS data. 
    We will need to choose either the + to -, or - to + curve and fit to moment. We will need to create a custom 
    mask to block out all data other than the fitting range, endpoint + 5-10% of data length should be sufficient 
    for a good fit. Code is setup to fit to negative field region. 

    1. Create mask array same shape as data.
    2. Mask (set value to 1) data for half of loop, I chose to mask + to - curve (everything up to half_loop).
    3. Mask everything in other curve (- to + half) other than the first 10% of data points. 
    4. Apply mask to PPMS data
    5. Setup linear fit
    """   

    mask_set = np.zeros(x_axis.shape) #Setup masking array same shape as array to be masked or similiar
    mask_set[:(half_loop)] = 1 #NP mask works by using 1 to mean mask this value so we add ones everywhere to 
    mask_set[(half_loop + int(half_loop*.1)):] = 1  #############
    #our zero array where we want to mask.
    fit_region = ma.masked_array(PPMS_data_red["MOKE Signal"], mask_set) #Apply masking portion

    #Linear fit of pre and post-edge.
    coeffs = np.ma.polyfit(x_axis, fit_region, 1)  #Linear fit coefficients. 
    bck_eqtn = np.poly1d(coeffs) #Turn coefficients into a fit equation
    bck_line = bck_eqtn(x_axis) #Apply x-values to fit equation for line values. 

    ax[0,1].set_title("Background Linear Fit")
    ax[0,1].plot(x_axis, bck_line,label = 'Fit')
    ax[0,1].plot(x_axis, PPMS_data_red["MOKE Signal"],label = 'Data')
    ax[0,1].grid()
    ax[0,1].set_xlabel("Voltage (V)")
    ax[0,1].set_ylabel("MOKE Signal")
    ax[0,1].legend()
    MH_bck_corr = PPMS_data_red["MOKE Signal"]

    ax[1,0].set_title("Background Corrected")
    ax[1,0].plot(x_axis, MH_bck_corr)
    ax[1,0].grid()
    ax[1,0].set_xlabel("Voltage (V)")
    ax[1,0].set_ylabel("MOKE Signal")
    #pdf.savefig(bbox_inches = "tight")

    #####################################################################################################


    ###Center y-axis around y = 0 and find +Ms and -Ms####


    """
    Now that MH loop is background corrected we need to find +Ms and -Ms so we can center the data about y = 0. 
    I think we can achieve this by averaging the first/last 5% of signal data points at either field endpoint. This will
    tell us by how much to shift up/down the data. 

    1. Center around y = 0 by finding y max/min and averaging to get the center value, subtract from moment data.
    """
    pMs = MH_bck_corr[:int(half_loop*0.05)].mean() #Average first 5% of points.
    nMs = MH_bck_corr[half_loop: (half_loop + int(half_loop*0.05)):].mean() #Average 5% of points at other endpoint.
    Ms_center = np.mean([pMs, nMs])



    ## for MH_bck_corr as the full data set and then put the conditions found for pMs and nMs as the conditions for 
    ##Data analysis to get rid of the outliers
    ##Would need to put in a if loop and output a new data frame. From that go ahead and do the verticle and horizontal 
    ##Corrections

    ax[1,1].set_title("Finding vertical center")
    ax[1,1].plot(x_axis, MH_bck_corr,label = 'MH')
    ax[1,1].hlines(pMs, x_axis[0],x_axis[half_loop], colors = 'seagreen', linestyles = 'dashed', label = "+Ms")
    ax[1,1].hlines(nMs, x_axis[0],x_axis[half_loop], colors = 'coral', linestyles = 'dashed', label = "-Ms")
    ax[1,1].hlines(Ms_center, x_axis[0], x_axis[half_loop], colors = 'blueviolet', linestyles = 'dashed', label = "Hori. Center")
    ax[1,1].grid()
    ax[1,1].set_xlabel("Voltage (V)")
    ax[1,1].set_ylabel("MOKE Signal")
    ax[1,1].legend()

    MH_vcent = MH_bck_corr - Ms_center #Center data vertically about moment = 0 emu
    MH_vcent_max= np.max(MH_vcent)
    
    MH_vcent_min=np.min(MH_vcent)
    MH_vcent_avg= (MH_vcent_max-MH_vcent_min)/2
    #print (MH_vcent_max,MH_vcent_min,MH_vcent_avg)
    
    fig2, ax2 = plt.subplots(1,2, figsize = (16,6))
    ax2[0].set_title("vertically centered around y = 0")
    ax2[0].plot(x_axis, MH_vcent)
    ax2[0].hlines(0, x_axis[0], x_axis[half_loop],  colors = 'black', linestyles = 'dashed')
    ax2[0].grid()
    ax2[0].set_xlabel("Voltage (V)")
    ax2[0].set_ylabel("MOKE Signal")
    ###########################################################################################################

########Center x-axis around x = 0 and find Coersive
    """
    Now we want to center the x-axis around field = 0 Oe
    1. Interpolate both halves of MH loop

    """
    
    data = np.array([x_axis, MH_vcent])
    data = data.T
    PPMS_data_blue=pd.DataFrame(data, columns=['Voltage (V)', 'MOKE Signal'])

    MHpn=PPMS_data_blue[:half_loop] # the first half +to-
    MHpn = MHpn.reset_index(drop = True)
    MHpn_inter=interp1d(y=MHpn['MOKE Signal'],x=MHpn['Voltage (V)'], kind='linear')
    
    
    MHnp=PPMS_data_blue[half_loop:] # the second half -to+
    MHnp = MHnp.reset_index(drop = True)
    MHnp_inter=interp1d(y=MHnp['MOKE Signal'], x=MHnp['Voltage (V)'], kind='linear')
                        
    
    int_ax = np.linspace(-1.7, 1.7, 2000)
    H_c_ind = np.absolute(MHnp_inter(int_ax)).argmin()
    zeropoint=int_ax[H_c_ind]
    # zeropoint=MHnp_inter(0)
    print ('zeropoint: ', zeropoint)
    
    H_c_ind = np.absolute(MHpn_inter(int_ax)).argmin()
    zeropoint2=int_ax[H_c_ind]
    # zeropoint=MHpn_inter(0)
    print('zeropoint2: ', zeropoint2)
    

    Deltazero=(zeropoint-zeropoint2)
    print (Deltazero)
    Coersive=abs(Deltazero)/2
    print ('Coercive field: ', Coersive)


    ax2[1].set_title("Finding Horizontal center")
    ax2[1].plot(x_axis, MH_vcent)
    ax2[1].vlines(zeropoint, MH_vcent[0],MH_vcent[half_loop], colors = 'green', linestyles = 'dashed', label = "+C")
    ax2[1].vlines(zeropoint2, MH_vcent[0],MH_vcent[half_loop],colors = 'coral', linestyles = 'dashed', label = "-C")
    ax2[1].vlines((zeropoint-Coersive), MH_vcent[0], MH_vcent[half_loop], colors = 'blueviolet', linestyles = 'dashed', label = "Hori. Center")
    ax2[1].set_xlabel("Voltage (V)")
    ax2[1].set_ylabel("MOKE Signal")
    ax2[1].grid()       

    MH_hcent=x_axis-(zeropoint-Coersive)


    fig3, ax3 = plt.subplots(1,2, figsize = (18,6))
    ax3[0].set_title("Horizontally centered around x = 0")
    ax3[0].plot(MH_hcent, MH_vcent)
    ax3[0].vlines(0, MH_vcent[0], MH_vcent[half_loop], colors = 'black', linestyles = 'dashed')
    ax3[0].set_xlabel("Voltage (V)")
    ax3[0].set_ylabel("MOKE Signal")
    ax3[0].grid()
    ax3[0].set_xlabel("Voltage (V)")
    #pdf.savefig(bbox_inches = "tight")
    
    
    ##Normalized data 
    # finding the max field value (MOKE data would be min field since start with neg )
    y_max=MH_vcent.max() 
    #print (y_max)
    y_min=MH_vcent.min()# this is the last field point in +field 
    #print  (y_min)
    y_avg=(y_max+(-1*y_min))/2
    #A=PPMS_data['Moment (emu/ccm)'].nsmallest(2)
    
    y_Nor = MH_vcent/ y_avg 
    
    ax3[1].set_title("20220816_Ta1_Pt1.2x2_Co1.2x2_Cu2_PolarMOKE_BackGlass")
    ax3[1].hlines(0, x_axis[0], x_axis[half_loop],  colors = 'black', linestyles = 'dashed')
    ax3[1].vlines(0, y_Nor[0], y_Nor[half_loop],  colors = 'black', linestyles = 'dashed')
    ax3[1].plot((MH_hcent*(5.44691536)/10), y_Nor)
    ax3[1].set_xlabel("Field (kG)")
    ax3[1].set_ylabel("Normalized MOKE Signal")
    ax3[1].grid()
    ax3[1].set_xlim(-2,2)
    
    coersive_T = abs((Coersive*(-1.3171)-0.0467)/10) #converting coercive field of V to T
    print('coercive (T) : ', coersive_T, 'Tesla')
```
